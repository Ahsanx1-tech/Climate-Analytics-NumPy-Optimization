**Pandas to NumPy Bridge**

```python
import pandas as pd
import numpy as np

df = pd.read_csv('Weather Dataset.csv')

weather_np = df.iloc[:, 1:7].to_numpy()

print(f"Array Shape: {weather_np.shape}")
```

**Multidimensional Slicing and Extraction**

```python
temp_c = weather_np[:, 0]
humidity = weather_np[:, 2]
wind_speed = weather_np[:, 3]
visibility = weather_np[:, 4]
```

**Vectorized Calculations**

```python
temp_f = (temp_c * 9/5) + 32

print(f"Mean Temperature in Fahrenheit: {np.mean(temp_f):.2f}°F")
```

**Boolean Masking**

```python
hazard_mask = (visibility < 5) & (wind_speed > 10)

hazard_data = weather_np[hazard_mask]

print(f"Hazardous instances detected: {len(hazard_data)}")
```

**Aggregation and Statistical Descriptives**

```python
summary = {
    "Max Temperature": np.max(temp_c),
    "Min Temperature": np.min(temp_c),
    "Avg Humidity": np.mean(humidity),
    "Pressure Stability (StdDev)": np.std(weather_np[:, 5])
}

for stat, val in summary.items():
    print(f"{stat}: {val:.2f}")
```

**Visualization**

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 3))
plt.plot(temp_c, label='Hourly Temperature', color='blue')
plt.axhline(y=np.mean(temp_c), color='orange', linestyle='--', label='Average')

plt.title("Temperature Trends")
plt.xlabel("Hours")
plt.ylabel("Celsius")
plt.legend()
plt.show()
```
